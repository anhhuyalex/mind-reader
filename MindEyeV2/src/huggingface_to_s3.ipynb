# Transfer huggingface mindeyev2 dataset to Stability aws s3

```python
import os
# from subprocess import call
# PS Note: it's faster to print the wget statements and then manually copy paste all them into terminal than to use subprocess call()
tmp = '/fsx/proj-fmri/shared/mindeyev2_dataset/' #'/scratch/mindeyev2_dataset/'

hf_base_link = 'https://huggingface.co/datasets/pscotti/mindeyev2/resolve/main/'

os.makedirs(tmp,exist_ok=True)

files = [
    "shared1000.npy",
    "betas_all_subj01.hdf5",
    "betas_all_subj02.hdf5",
    "betas_all_subj03.hdf5",
    "betas_all_subj04.hdf5",
    "betas_all_subj05.hdf5",
    "betas_all_subj06.hdf5",
    "betas_all_subj07.hdf5",
    "betas_all_subj08.hdf5",
    "coco_images_224_float16.hdf5",
    "COCO_73k_subj_indices.hdf5",
]

for f in files:    
    command = f"wget --show-progress {hf_base_link}{f} -O {tmp}{f}"
    print(command)
    # call(command,shell=True)

for sub in range(1,9):
    subject = f'subj0{sub}'

    tmp_fol = f'{tmp}wds/{subject}/'
    os.makedirs(tmp_fol,exist_ok=True)
    os.makedirs(tmp_fol+'train',exist_ok=True)
    os.makedirs(tmp_fol+'test',exist_ok=True)

    for i in range(37):
        link = f'train/{i}.tar'
        command = f"wget --show-progress {hf_base_link}wds/{subject}/{link} -O {tmp}wds/{subject}/{link}"
        print(command)
        # call(command,shell=True)

    link = f'test/0.tar'
    command = f"wget --show-progress {hf_base_link}wds/{subject}/{link} -O {tmp}wds/{subject}/{link}"
    print(command)
    # call(command,shell=True)

command = "aws s3 sync /scratch/mindeyev2_dataset s3://proj-fmri/mindeyev2_dataset --region us-west-2"
print(command)
```
